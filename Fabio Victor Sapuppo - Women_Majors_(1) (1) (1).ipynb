```python
import pandas as pd
%matplotlib inline
direct_link = 'http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv'
women_majors = pd.read_csv(direct_link)

women_majors = women_majors.rename(columns={'Year':'Ano','Art and Performance':'Arte e Performance','Education':'Educação','English':'Ingles','Foreign Languages':'Linguas Estrangeiras','Health Professions':'Profissoes da Saude','Public Administration':'Administração Publica'})

print(women_majors.info())
women_majors.head()
```

```python
# Abaixo de 20%
#under_20 = women_majors.loc[0, women_majors.loc[0]<20]
#under_20
over_50 = women_majors.loc[0, (women_majors.loc[0]>50) & (women_majors.loc[0]<100)]
over_50
```
